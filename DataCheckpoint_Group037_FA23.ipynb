## Dataset GDP of Countries Over Years


```python
path = "/content/drive/MyDrive/country_gdp.csv"
GDP = pd.read_csv(path, sep=';')
GDP
```

```python
shape_GDP = GDP.shape
shape_GDP
```

```python
GDP = GDP.iloc[:, :4].join(GDP.iloc[:, -4:])
GDP = GDP.iloc[:, :-1]
```

```python
GDP.drop(['Indicator Name', 'Indicator Code'], axis=1, inplace=True)
GDP
```

```python
GDP = GDP.dropna(subset=['2020', '2021', '2022'], how='any')
```

```python
GDP
```

##Putting it together: Matching both datasets with ISO codes from covid_cleaned and Country Code from GDP


```python
iso_codes_df1 = covid_cleaned['iso_code'].unique()
iso_codes_df2 = GDP['Country Code'].unique()

# Find common ISO codes
common_iso_codes = set(iso_codes_df1) & set(iso_codes_df2)

# Filter rows in both datasets based on common ISO codes
filtered_covid = covid_cleaned[covid_cleaned['iso_code'].isin(common_iso_codes)]
filtered_GDP = GDP[GDP['Country Code'].isin(common_iso_codes)]
```
