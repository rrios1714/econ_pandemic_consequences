##Exploring the relation of GDP to total Covid-19 Cases

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

pd.options.mode.chained_assignment = None  # Default='warn'

# Assuming 'Country Code' is the common identifier between the two datasets
merged_data = pd.merge(filtered_GDP, filtered_covid, left_on='Country Code', right_on='iso_code')

# Function to create lmplot with linear regression for a specific year
def create_lmplot(year_data, year):
    # Convert the 'date' column to datetime format
    year_data['date'] = pd.to_datetime(year_data['date'])

    # Filter the DataFrame for the specified date (December 30, of the given year)
    end_year_data = year_data[year_data['date'] == f'{year}-12-30']

    # Set a seaborn style
    sns.set(style="whitegrid")

    # Create an lmplot with linear regression for the specific year
    plt.figure(figsize=(12, 8))
    sns.regplot(x=f'{year}', y='total_cases', data=end_year_data, scatter_kws={"s": 100, "alpha": 0.8})

    correlation_coefficient, _ = pearsonr(end_year_data[f'{year}'], end_year_data['total_cases'])
    print(f"Correlation Coefficient for {year}: {correlation_coefficient}")

    # Set labels and title
    plt.xlabel(f'GDP in {year}')
    plt.ylabel('Total COVID-19 Cases')
    plt.title(f'Regression Plot of GDP ({year}) vs. Total COVID-19 Cases (Dec 30, {year})')

    # Show plot
    plt.show()

# Create regression plots for 2020, 2021, and 2022
create_lmplot(merged_data[merged_data['date'].str.startswith('2020')], 2020)
create_lmplot(merged_data[merged_data['date'].str.startswith('2021')], 2021)
create_lmplot(merged_data[merged_data['date'].str.startswith('2022')], 2022)
```
